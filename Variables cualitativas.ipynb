```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

random.seed(42)

nivel_estudios = []

nivel_estudios.extend(['Secundaria'] * 110)
nivel_estudios.extend(['Bachillerato'] * 172)
nivel_estudios.extend(['Grado'] * 145)
nivel_estudios.extend(['Master'] * 52)
nivel_estudios.extend(['Doctorado'] * 21)

nivel_estudios = np.random.permutation(nivel_estudios)

df = pd.DataFrame({
    'Nivel_Estudios': nivel_estudios
})

print(df.head())
```

```python
#Contar la frecucuencia de cada categoria
frecuencia = df['Nivel_Estudios'].value_counts()

print(frecuencia)
```

```python
#ordenar las categorias
orden = ['Secundaria', 'Bachillerato', 'Grado', 'Master', 'Doctorado']
frecuencia = frecuencia.reindex(orden)

print(frecuencia)
```

```python
#diagrama de barras
plt.figure(figsize=(8,5))

frecuencia.plot(kind='bar')

plt.title('Diagrama de barras')
plt.xlabel('Nivel de estudios')
plt.ylabel('Frecuencia absoluta')

plt.show()
```

```python
#diagrama pastel
plt.figure(figsize=(7,7))

frecuencia.plot(
    kind='pie',
    autopct='%1.2f%%',
    startangle=90
)

plt.title('Diagrama de sectores')
plt.ylabel('')

plt.show()
```
